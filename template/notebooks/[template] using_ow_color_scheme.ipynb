# Line plot

Two random walks

```python
import random

walk_1 = [0]
walk_2 = [0]
steps = 1000

for i in range(steps):
    change = 1 if random.randint(0, 1) else -1
    walk_1.append(walk_1[i] + change)
    walk_2.append(walk_2[i] + np.random.normal())
    
plt.plot(walk_1)
plt.plot(walk_2)
plt.show()
```

# Scatter plot

This uses a color map based on clusters labels assigned to artificially created clusters.

```python
blobs_data = []
N = 500

for _ in range(4):

    # Create artificial clusters
    blobs = np.concatenate([np.c_[(np.random.multivariate_normal([20*np.random.rand(), 20*np.random.rand()], 
                                   [[(np.random.rand()+1)/2, 0], [0, (np.random.rand()+1)/2]], size=N), 
                                   np.full(N, i))] for i in range(3)])
    blobs_data.append(blobs)
    
fig, ax = plt.subplots(2, 2, figsize=(8, 8))
for idx, blobs in enumerate(blobs_data):
    x, y = idx // 2, idx % 2
    X, Y, clusters = blobs[:, 0], blobs[:, 1], blobs[:, 2]
    ax[x, y].scatter(X, Y, c=clusters, cmap=ow_colormap());

plt.show()
```

**We have created various colormaps using OW colors**

You can access this through arguments for `ow_colormap()`

```python
fig, ax = plt.subplots(2, 2, figsize=(8, 8))
for idx, blobs in enumerate(blobs_data):
    x, y = idx // 2, idx % 2
    X, Y, clusters = blobs[:, 0], blobs[:, 1], blobs[:, 2]
    ax[x, y].scatter(X, Y, c=clusters, cmap=ow_colormap("greys"));

plt.show()
```
